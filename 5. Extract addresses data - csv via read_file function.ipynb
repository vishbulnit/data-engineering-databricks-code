# Extract data from the adress files
1. Demonstrate limitations of csv file fromat in select statement
2. Use read_files function to overcome the limitations
3. Create addressess view in the bronze layer

### 1. Demonstrate limitations of csv file fromat in select statement

```python
%sql
select * from csv.`/Volumes/gizmobox/landing/operations/addresses/` limit 5
```

### 2. Use read_files function to overcome the limitations

```python
%sql
select * from read_files('/Volumes/gizmobox/landing/operations/addresses/',
                        format=>'csv',
                        header=>True,
                        delimiter=>'\t')
                        limit 5;
```

### 3. Create addressess view in the bronze layer

```python
%sql
create or replace view gizmobox.bronze.v_addresses
as
select * from read_files('/Volumes/gizmobox/landing/operations/addresses/',
                        format=>'csv',
                        header=>True,
                        delimiter=>'\t');
```
