# Extract data - using pyspark
1. Run sql commands using python - spark.sql function
2. Spark dataframe reader api
3. Read tables using spark.table function

#### 1. Run sql commands using python - spark.sql function


```python
df = spark.sql('select * from json.`/Volumes/gizmobox/landing/operations/customers/`')
df.show(5)
```

```python
df1 = spark.sql('''
                create or replace temp view tv_customers 
                as
                select * from json.`/Volumes/gizmobox/landing/operations/customers/`
                ''');
```

```python
%sql
select * from tv_customers limit 5
```

#### 2. Spark dataframe reader api

```python
df2 = spark.read.json('/Volumes/gizmobox/landing/operations/customers/')
```

```python
df2.show(5)
```

```python
df3 = spark.read.format('json').load('/Volumes/gizmobox/landing/operations/customers/')
df3.show(5)
```
