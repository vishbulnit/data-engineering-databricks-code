## Extract Data From the Payments Files
1. Read Payments File
1. Create Payments Table in Bronze Schema

### 1. List the files from payment folder

```python
%fs ls 'abfss://gizmobox@datastorageexternalcon.dfs.core.windows.net/landing/external_data/payments'
```

### 2. Read the payments file

```python
payments_schema = 'payment_id INTEGER, order_id INTEGER, payment_timestamp TIMESTAMP, payment_status INTEGER, payment_method STRING'
```

```python
df = (
    spark.read.format('csv')
         .option('delimiter', ',')
         .schema(payments_schema)
         .load('abfss://gizmobox@datastorageexternalcon.dfs.core.windows.net/landing/external_data/payments')
)
display(df)
```

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, TimestampType

py_payments_schema = StructType([
    StructField("payment_id", IntegerType()),
    StructField("order_id", IntegerType()),
    StructField("payment_timestamp", TimestampType()),
    StructField("payment_status", IntegerType()),
    StructField("payment_method", StringType())
])
```

```python
df = (
    spark.read.format('csv')
         .option('delimiter', ',')
         .schema(py_payments_schema)
         .load('abfss://gizmobox@datastorageexternalcon.dfs.core.windows.net/landing/external_data/payments')
)
display(df)
```

### 3. Create Payment Table in the bronze schema

```python
df.writeTo('gizmobox.bronze.py_payments').createOrReplace()
```
