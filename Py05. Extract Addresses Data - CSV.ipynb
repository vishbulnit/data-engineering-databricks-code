## Extract Data From the Address Files
1. Read all the addresses files
1. Create addresses table in the bronze layer

### 1. Read all the address files

```python
df = (spark.read.format("csv")
    .option("header", "true")
    .option("delimiter", "\t")
    .load("/Volumes/gizmobox/landing/operations/addresses/")
)
display(df)
```

### 2. Create Addresses table in the Bronze Layer

```python
df.writeTo('gizmobox.bronze.py_addresses').createOrReplace()
```
