# Transform customer data
1. Remove records with null customer id
2. Remove exact diplicate records
3. Remove duplicate records based on created_timestamp
4. Cast the column to the correct data type
5. Write transformed data to the silver schema

#### 1. Remove records with null customer id


```python
%sql
select * from gizmobox.bronze.v_customer
where customer_id is not null;
```

#### 2. Remove exact diplicate records


```python
%sql
select * from gizmobox.bronze.v_customer
where customer_id is not null
order by customer_id ;
```

#### 3. Remove duplicate records based on created_timestamp


```python
%sql
select
distinct * 
from gizmobox.bronze.v_customer
where customer_id is not null
order by customer_id ;
```

```python
%sql
select
customer_id,
max(created_timestamp),
max(customer_name),
max(date_of_birth),
max(email),
max(member_since),
max(telephone)
from gizmobox.bronze.v_customer
where customer_id is not null
group by customer_id
order by customer_id ;
```

```python
%sql
create or replace temporary view v_customer_distinct
as
select 
distinct * from gizmobox.bronze.v_customer
where customer_id is not null
order by customer_id ;
```

```python
%sql
select
customer_id,
MAX(created_timestamp) as max_created_timestamp
FROM v_customer_distinct
group by customer_id;
```

```python
%sql
with cte_max
as 
(
select
customer_id,
MAX(created_timestamp) as max_created_timestamp
FROM v_customer_distinct
group by customer_id
)
select 
f.*
from v_customer_distinct as f
join cte_max as t
on f.customer_id = t.customer_id
and f.created_timestamp = t.max_created_timestamp;

```

#### 4. Cast the column to the correct data type

```python
%sql
with cte_max
as 
(
select
customer_id,
MAX(created_timestamp) as max_created_timestamp
FROM v_customer_distinct
group by customer_id
)
select 
cast(f.created_timestamp as timestamp) as created_timestamp,
f.customer_id,
f.customer_name,
cast(f.date_of_birth as date) as date_of_birth,
f.email,
cast(f.member_since as date) as member_since,
f.telephone
from v_customer_distinct as f
join cte_max as t
on f.customer_id = t.customer_id
and f.created_timestamp = t.max_created_timestamp;
```

#### 5. Write transformed data to the silver schema

```python
%sql
create table gizmobox.silver.customer
as
with cte_max
as 
(
select
customer_id,
MAX(created_timestamp) as max_created_timestamp
FROM v_customer_distinct
group by customer_id
)
select 
cast(f.created_timestamp as timestamp) as created_timestamp,
f.customer_id,
f.customer_name,
cast(f.date_of_birth as date) as date_of_birth,
f.email,
cast(f.member_since as date) as member_since,
f.telephone
from v_customer_distinct as f
join cte_max as t
on f.customer_id = t.customer_id
and f.created_timestamp = t.max_created_timestamp;
```

```python
%sql
select * from gizmobox.silver.customer;
```

```python
%sql
DESCRIBE EXTENDED gizmobox.silver.customer;
```
