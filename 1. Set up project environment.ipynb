### 2. create external location for Gizmobox

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS databrics_course_ext_con_gizmobox
    URL 'abfss://gizmobox@datastorageexternalcon.dfs.core.windows.net/'
    WITH (STORAGE CREDENTIAL databrics_external_credential)
    COMMENT 'External location for GizmoBox data lakehouse'
```

### 3. Create the catalog - gizmobox

```python
%sql
SHOW CATALOGS;
```

```python
%sql
CREATE CATALOG IF NOT EXISTS gizmobox
    MANAGED LOCATION 'abfss://gizmobox@datastorageexternalcon.dfs.core.windows.net/'
    COMMENT 'This is the Catalog for GizmoBox data lakehouse'
```

### 4.Create Schemas

1. Landing
2. Bronze
3. Silver
4. Gold


```python
%sql
select current_catalog();
```

```python
%sql
use catalog gizmobox;
create schema if not exists landing
    managed location 'abfss://gizmobox@datastorageexternalcon.dfs.core.windows.net/landing';
create schema if not exists bronze
    managed location 'abfss://gizmobox@datastorageexternalcon.dfs.core.windows.net/landing';
create schema if not exists silver
    managed location 'abfss://gizmobox@datastorageexternalcon.dfs.core.windows.net/landing';
create schema if not exists gold
    managed location 'abfss://gizmobox@datastorageexternalcon.dfs.core.windows.net/landing';
```

```python
%sql
select current_catalog()
```

### 5. Create Volume

```python
%sql
use catalog gizmobox;
use schema landing;
create external volume if not exists operations
    LOCATION 'abfss://gizmobox@datastorageexternalcon.dfs.core.windows.net/landing/operational_data/';
```
