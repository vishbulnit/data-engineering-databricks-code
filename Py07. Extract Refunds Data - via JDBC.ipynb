## Extract Data From the External data
1. Read Return data
2. Create Returns Table in Bronze Schema

### 1. Read Returns Data

```python
%fs ls 'abfss://gizmobox@datastorageexternalcon.dfs.core.windows.net/landing/external_data/refunds/refund.txt'
```

```python
df = (
      spark.read.format('csv')
      .option("header", "true")
      .option("inferSchema", "true")
      .option("delimiter", ",")
      .load("abfss://gizmobox@datastorageexternalcon.dfs.core.windows.net/landing/external_data/refunds/refund.txt")
)
display(df)
```

### 2. Write the refunds data to the bronze schema

```python
df.writeTo('gizmobox.bronze.py_refunds').createOrReplace()
```

```python
%sql
SELECT * FROM gizmobox.bronze.py_refunds;
```
